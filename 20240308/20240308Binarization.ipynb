## Binarization

```python
# 載入套件
import matplotlib.pyplot as plt
import numpy as np
import cv2

# 讀取照片
img1 = cv2.imread("sena.jpg")
img2 = cv2.imread("moon.jpg")

# 利用list放一起，方便後面的for迴圈
imgs = [img1, img2]
titles = ['img1', 'img2']

# 利用for迴圈，逐一呈現
for i in range(2):
    # 秀出子圖
    plt.subplot(1,2,i+1)
    # 呈現灰階的圖片
    plt.imshow(imgs[i], 'gray')
    plt.title(titles[i])
    # 將X，Y軸的刻度省略
    plt.xticks([]),plt.yticks([])
```

```python
def general(image):
    image[np.where(image<128)] = 0
    image[np.where(image>127)] = 255
    return image
```

```python
image1_g = general(img1)
image2_g = general(img2)
images_g = [image1_g, image2_g]
titles = ['image1_g', 'image2_g']
# 利用for迴圈，逐一呈現
for i in range(2):
    # 秀出子圖
    plt.subplot(1,2,i+1)
    # 呈現灰階的圖片
    plt.imshow(images_g[i], 'gray')
    plt.title(titles[i])
    # 將X，Y軸的刻度省略
    plt.xticks([]),plt.yticks([])
```

### Inverse
### grey image(self-coding)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("sena.jpg")

images = [image, 255 - image]
titles = ['ORIGIN', 'INVERSE']

for i in range(2):
    # 秀出子圖
    plt.subplot(1,2,i+1)
    # 呈現灰階的圖片
    plt.imshow(images[i])
    plt.title(titles[i])
    # 將X，Y軸的刻度省略
    plt.xticks([]),plt.yticks([])
plt.show()
```
