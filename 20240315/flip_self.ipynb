```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread("Images/lena_color.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(image)
plt.show()
```

```python
image_g = image[:,:,0]*0.299+image[:,:,1]*0.587+image[:,:,2]*0.114
print(image_g.shape)
plt.figure()
plt.imshow(image_g,cmap = "gray")
plt.show()
```

```python
height, width = image_g.shape
img_lr = image_g.copy()
for i in range(height):
   for j in range(width):
      img_lr[i, j] = image_g[i,width-j-1]
plt.figure()
plt.imshow(img_lr,cmap = "gray")
plt.show() 
```
