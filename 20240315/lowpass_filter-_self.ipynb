```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#image = cv2.imread('images/lena_gray_noise_median.jpg', 0) #cv2.IMREAD_GRAYSCALE=0
image = cv2.imread('lena_gray_noise_lowpass.jpg', 0)
lowpass_image = image.copy()
height, width = image.shape
a = 1 #neighsor_height_distance
b = 1 #neighsor_width_distance
h = 2 * a + 1 #kernel_height
w = 2 * b + 1 #kernel_width
area = h * w #kernel_area
kernel = np.ones((h, w), np.float32)/area
print (kernel)
for i in range(a, height - a):
    for j in range(b, width - b):
        # arr's size is hxw = 5x5 (if a = b = 2)
        sum = 0
        for ki in range(-a,a+1):
            for kj in range(-b,b+1):
                sum = sum + image[i+ki, j+kj]*kernel[ki,kj]       
        lowpass_image[i, j] = int(sum)

images = [image, lowpass_image]
titles = ['ORIGINAL', 'LOWPASS']
for i in range(2):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
