```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread("images/sena.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print("image.shape = ", image.shape)
width, height = image.shape
reshaped_image = image.reshape(width*height)
#image.reshape((width*height)將image 拉成width*height維向量
print("reshaped_image.shape = ",reshaped_image.shape)
# 繪製直方圖(長條圖),圖的x和y軸分別表示bin值和每個bin的frequency)
plt.hist(reshaped_image, bins=256, color='skyblue', alpha=0.8) #alpha為透明度
#設置圖表屬性
plt.title('hist for image')
plt.xlabel('Gray Value')
plt.ylabel('Frequency')
#顯示圖表
plt.show()
```
