## 5. 對影像 highway.jpg 的 RGB 三個頻道分別畫出長條圖，顏色分別用 red, green, blue。此外，R channel 在數值 102 的長條片(bar or strip)，G channel 在數值 95 的長條片，B channel 在數值 87 的長條片都以黑色呈現。三個長條圖依序分別畫(呈現)於三列。輸出形式如下圖(注意這是採不同輸入影像的結果):


```python
# 下載套件
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. 讀影像 highway.jpg 並畫出(呈現)影像。
print("1. 讀影像 highway.jpg 並畫出(呈現)影像。")

image = cv2.imread('highway.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(image)
plt.xticks([]),plt.yticks([])
plt.show()

# 2. 對影像 highway.jpg 的 RGB 三個頻道分別畫出長條圖，顏色分別用 red, green, blue，依序畫(呈現)於同一列。
print("2. 對影像 highway.jpg 的 RGB 三個頻道分別畫出長條圖，顏色分別用 red, green, blue，依序畫(呈現)於同一列。")

bimg, gimg, rimg = cv2.split(image)
width, height, thickness = image.shape
reshaped_rimg = rimg.reshape(width*height)
reshaped_gimg = gimg.reshape(width*height)
reshaped_bimg = bimg.reshape(width*height)
reshaped_images = [reshaped_rimg, reshaped_gimg, reshaped_bimg]
titles = ['red', 'green', 'blue']
colors = ('r', 'g', 'b')

plt.figure(figsize=(24, 8))
for i, col in enumerate(colors):
    plt.subplot(1, 3, i+1)
    plt.hist(reshaped_images[i], bins=256, color=col)
    plt.title(titles[i])
plt.show()

# 3. 對影像 highway.jpg 的 RGB 三個頻道分別畫出分布圖，依序畫(呈現)於同一列，顏色分別用 red, green, blue。
print("3. 對影像 highway.jpg 的 RGB 三個頻道分別畫出分布圖，依序畫(呈現)於同一列，顏色分別用 red, green, blue。")

images = [rimg, gimg, bimg]

plt.figure(figsize=(24, 8))
for i, col in enumerate(colors):
    hist = cv2.calcHist([images[i]],[0],None,[256],[0,256])
    plt.subplot(1, 3, i+1)
    plt.plot(hist, color=col)
    plt.title(titles[i])
plt.show()

# 4. 請列印出 R channel 在數值 102 的頻率，G channel 在數值 95 的頻率，B channel 在數值 87 的頻率。
# 5. 對影像 highway.jpg 的 RGB 三個頻道分別畫出長條圖，顏色分別用 red, green, blue。
#    此外，R channel 在數值 102 的長條片(bar or strip)，G channel 在數值 95 的長條片，B channel 在數值 87 的長條片都以黑色呈現。
#    三個長條圖依序分別畫(呈現)於三列。
print("4. 請列印出 R channel 在數值 102 的頻率，G channel 在數值 95 的頻率，B channel 在數值 87 的頻率。\n")

freq_r, bin_edge_r, patches_r = plt.hist(reshaped_rimg, bins=256, color='r')
patches_r[102].set_facecolor('black')
print(f"R channel 在數值 102 的頻率 = {freq_r[102]}")

plt.figure()
freq_g, bin_edge_g, patches_g = plt.hist(reshaped_gimg, bins=256, color='g')
patches_g[95].set_facecolor('black')
print(f"G channel 在數值 95 的頻率  = {freq_g[95]}")

plt.figure()
freq_b, bin_edge_b, patches_b = plt.hist(reshaped_bimg, bins=256, color='b')
patches_b[87].set_facecolor('black')
print(f"B channel 在數值 87 的頻率  = {freq_b[87]}")

print("\n5. 對影像 highway.jpg 的 RGB 三個頻道分別畫出長條圖，顏色分別用 red, green, blue。此外，R channel 在數值 102 的長條片(bar or strip)，G channel 在數值 95 的長條片，B channel 在數值 87 的長條片都以黑色呈現。三個長條圖依序分別畫(呈現)於三列。")
plt.show()


# 用for迴圈印出RGB三個channel的圖
# for i, col in enumerate(colors):
#     plt.figure()
#     freq, bin_edge, patches = plt.hist(reshaped_images[i], bins=256, color=colors[i])
```
