```python
# 
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1.  highway.jpg ()
print("1.  highway.jpg ()")

image = cv2.imread('highway.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure()
plt.imshow(image)
plt.xticks([]),plt.yticks([])
plt.show()

# 2.  highway.jpg  RGB  red, green, blue()
print("2.  highway.jpg  RGB  red, green, blue()")

bimg, gimg, rimg = cv2.split(image)
width, height, thickness = image.shape
reshaped_rimg = rimg.reshape(width*height)
reshaped_gimg = gimg.reshape(width*height)
reshaped_bimg = bimg.reshape(width*height)
reshaped_images = [reshaped_rimg, reshaped_gimg, reshaped_bimg]
titles = ['red', 'green', 'blue']
colors = ('r', 'g', 'b')

plt.figure(figsize=(24, 8))
for i, col in enumerate(colors):
    plt.subplot(1, 3, i+1)
    plt.hist(reshaped_images[i], bins=256, color=col)
    plt.title(titles[i])
plt.show()

# 3.  highway.jpg  RGB () red, green, blue
print("3.  highway.jpg  RGB () red, green, blue")

images = [rimg, gimg, bimg]

plt.figure(figsize=(24, 8))
for i, col in enumerate(colors):
    hist = cv2.calcHist([images[i]],[0],None,[256],[0,256])
    plt.subplot(1, 3, i+1)
    plt.plot(hist, color=col)
    plt.title(titles[i])
plt.show()

# 4.  R channel  102 G channel  95 B channel  87 
# 5.  highway.jpg  RGB  red, green, blue
#    R channel  102 (bar or strip)G channel  95 B channel  87 
#    ()
print("4.  R channel  102 G channel  95 B channel  87 \n")

freq_r, bin_edge_r, patches_r = plt.hist(reshaped_rimg, bins=256, color='r')
patches_r[102].set_facecolor('black')
print(f"R channel  102  = {freq_r[102]}")

plt.figure()
freq_g, bin_edge_g, patches_g = plt.hist(reshaped_gimg, bins=256, color='g')
patches_g[95].set_facecolor('black')
print(f"G channel  95   = {freq_g[95]}")

plt.figure()
freq_b, bin_edge_b, patches_b = plt.hist(reshaped_bimg, bins=256, color='b')
patches_b[87].set_facecolor('black')
print(f"B channel  87   = {freq_b[87]}")

print("\n5.  highway.jpg  RGB  red, green, blueR channel  102 (bar or strip)G channel  95 B channel  87 ()")
plt.show()


# forRGBchannel
# for i, col in enumerate(colors):
#     plt.figure()
#     freq, bin_edge, patches = plt.hist(reshaped_images[i], bins=256, color=colors[i])
```
