## 2. 對影像 highway.jpg 的 RGB 三個頻道分別畫出長條圖,顏色分別用 red, green, blue,依序畫(呈現)於同一列,輸出形式如下圖(注意這是採不同輸入影像的結果)


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("highway.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

bimg, gimg, rimg = cv2.split(img) # 分離 B、G、R 通道
#bimg, gimg, rimg = img[:,:,0], img[:,:,1], img[:,:,2]
width, height, thickness = img.shape
reshaped_rimg = rimg.reshape(width*height)
reshaped_gimg = gimg.reshape(width*height)
reshaped_bimg = bimg.reshape(width*height)
images = [reshaped_rimg, reshaped_gimg, reshaped_bimg]
titles = ['red', 'green', 'blue']
colors = ('r','g', 'b')
for i, col in enumerate(colors):
  plt.subplot(1,3,i+1)
  plt.hist(images[i], bins=256, color=col)
  plt.title(titles[i])
  plt.xticks([]),plt.yticks([])
plt.show()
```

## 4. 請列印出 R channel 在數值 102 的頻率, G channel 在數值 95 的頻率, B channel 在數值 87 的頻率

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("highway.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


bimg, gimg, rimg = cv2.split(img)

# 使用列表來存儲各通道的圖像和顏色標籤
images = [rimg, gimg, bimg]
colors = ('r', 'g', 'b')
labels = ("red", "green", "blue")
target_values = [102, 95, 87]  # 分別對應 R、G、B 通道的目標數值

# 計算並顯示直方圖
for i, color in enumerate(colors):
    hist = cv2.calcHist([images[i]], [0], None, [256], [0,256])
    plt.subplot(1, 3, i+1)
    plt.plot(hist, color=color)
    plt.title(f"{labels[i]} channel")

    # 取得並列印特定數值的頻率
    freq = hist[target_values[i]][0]
    print(f"{labels[i]} channel at value {target_values[i]} has frequency: {freq}")

plt.show()
```
