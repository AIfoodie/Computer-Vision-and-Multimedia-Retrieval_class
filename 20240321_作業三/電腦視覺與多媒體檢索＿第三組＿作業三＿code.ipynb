### 3. 對影像 highway.jpg 做 gray scaling 處理後, 命名為 highway_gray.jpg 的圖檔

```python
# 載入套件
import matplotlib.pyplot as plt
import numpy as np
import cv2

# 讀取照片
img = cv2.imread("highway.jpg")

# 將彩色圖片轉換為灰階圖片
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 儲存灰階圖片
cv2.imwrite("highway_gray.jpg", gray_img)

# 呈現灰階圖片
plt.imshow(gray_img, cmap='gray')
plt.title('Gray Image')

# 將X，Y軸的刻度省略
plt.xticks([]),plt.yticks([])

# 顯示圖片
plt.show()
```

### 4. 對影像 highway.jpg 做 high pass filtering 處理後, 命名為 highway_highpass.jpg 的圖檔

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('highway.jpg', 0)
highpass_image = image.copy()
height, width = image.shape

a = 1 #neighsor_height_distance
b = 1 #neighsor_width_distance
h = 2 * a + 1 #kernel_height
w = 2 * b + 1 #kernel_width
area = h * w #kernel_area

#kernel = np.ones((h, w), np.float32)/area
kernel = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
print (kernel)

for i in range(a, height - a):
    for j in range(b, width - b):
        # arr's size is hxw = 5x5 (if a = b = 2)
        sum = 0
        for ki in range(-a,a+1):
            for kj in range(-b,b+1):
                sum = sum + image[i+ki, j+kj]*kernel[ki,kj]       
        highpass_image[i, j] = int(sum)

# 保存影像
cv2.imwrite("highway_highpass.jpg", highpass_image)

# 原始影像和濾波後的影像呈現
plt.figure(figsize=(20, 10)) # 圖的size
plt.subplot(1, 2, 1) # 該圖的位置
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis() # 呈現 X，Y軸的刻度

plt.subplot(1, 2, 2) # 該圖的位置
plt.imshow(highpass_image, cmap='gray')
plt.title('High Pass Filtered Image')
plt.axis('off') # 將X，Y軸的刻度省略

plt.show()
```

### 5. 將以下三張影像依序畫(呈現)於同一列: highway.jp, highway_gray.jpg, highway_highpass.jpg

```python
import matplotlib.pyplot as plt
import cv2

# 讀取三張影像
img_color = cv2.imread("highway.jpg")
img_gray = cv2.imread("highway_gray.jpg")
img_highpass = cv2.imread("highway_highpass.jpg")

# 設定圖片大小
plt.figure(figsize=(18, 6))

# 顯示彩色影像
plt.subplot(1, 3, 1) # 同一列的第一行
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# 顯示灰階影像
plt.subplot(1, 3, 2) # 同一列的第二行
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
plt.title('Gray Scale Image')
plt.axis('off')

# 顯示高通濾波後影像
plt.subplot(1, 3, 3) # 同一列的第三行
plt.imshow(cv2.cvtColor(img_highpass, cv2.COLOR_BGR2RGB))
plt.title('High Pass Filtered Image')
plt.axis('off')

# 顯示圖片
plt.show()
```
